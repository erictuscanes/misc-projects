```python
import pandas as pd
import os
import numpy as np
```

```python
mc_file = os.path.join("MC 2018 Production through Q3.csv")
mc_df = pd.read_csv(mc_file)
```

```python
mc_df.head()
```

```python
estimate_count = len(mc_df["Approved Estimate"].value_counts())

total_estimates = pd.DataFrame({"Total Estimates":[estimate_count]})
total_estimates
```

```python
brand_estimate = mc_df[["Brand", "Approved Estimate"]]
brand_estimate.head()
```

```python
agency_estimate = mc_df[["Agency", "Approved Estimate"]]
agency_estimate.head()
```

```python
media_estimate = mc_df[["Media Type", "Approved Estimate"]]
media_estimate.head()
```

```python
production_estimate = mc_df[["Production Type", "Approved Estimate"]]
production_estimate.head()
```

```python
# Create the GroupBy object based on the "Type 1" column
brand_group = brand_estimate.groupby(["Brand"])

brand_mean = (brand_group["Approved Estimate"].mean())

brand_sum = (brand_group["Approved Estimate"].sum())

brand_count = (brand_group["Approved Estimate"].count())

brand_analysis = pd.DataFrame ({
    "Number of Shoots": brand_count,
    "Average Purchase Price": brand_mean.map('${:,.2f}'.format),
    "Total Purchase": brand_sum.map('${:,.2f}'.format),
})

brand_analysis
```

```python
# Create the GroupBy object based on the "Type 1" column
agency_group = agency_estimate.groupby(["Agency"])

agency_mean = (agency_group["Approved Estimate"].mean())

agency_sum = (agency_group["Approved Estimate"].sum())

agency_count = (agency_group["Approved Estimate"].count())

agency_analysis = pd.DataFrame ({
    "Number of Shoots": agency_count,
    "Average Purchase Price": agency_mean.map('${:,.2f}'.format),
    "Total Purchase": agency_sum.map('${:,.2f}'.format),
})

agency_analysis
```
