```python
import pandas as pd
import os
import numpy as np
```

```python
mc_file = os.path.join("MC 2018 Production through Q3.csv")
mc_df = pd.read_csv(mc_file)
```

```python
mc_df.head()
```

```python
brand_estimate = mc_df[["Brand", "Approved Estimate"]]
brand_estimate.head()
```

```python
agency_estimate = mc_df[["Agency", "Approved Estimate"]]
agency_estimate.head()
```

```python
media_estimate = mc_df[["Media Type", "Approved Estimate"]]
media_estimate.head()
```

```python
production_estimate = mc_df[["Production Type", "Approved Estimate"]]
production_estimate.head()
```

```python
# Create the GroupBy object based on the "Type 1" column
brand_group = brand_estimate.groupby(["Brand"])

brand_type = brand_group.mean()
brand_type.style.format({"Approved Estimate":"${:,.2f}"})
```

```python
brand_sum = brand_group.sum()
brand_sum
brand_sum.style.format({"Approved Estimate":"${:,.2f}"})
```

```python
brand_value = brand_group.count()
brand_value
```
